# Normal Distribution: Complete Overview 

Definition 

The normal distribution,also known as the Gaussian distribution, is a continuous probability distribution characterized by its symmetric, bell-shaped curve. It is defined by two parameters: the mean (Œº) and the standard deviation (ùúé).


### Key terms for Normal Distribution 

1. **Error**: The difference between a data point and a predicted or average value. 
2. **Standardize**:  Subtract the mean and divide by the standard deviation. 
3. **z-score**:  The result of standardizing an individual data point. Standard normal A normal distribution with mean = 0 and standard deviation = 1. 
4. **QQ-Plot**: A plot to visualize how close a sample distribution is to a specified distribution, e.g., the normal distribution


## Properties of Normal Distribution 

1. **Symmetry**:

- The normal distribution is symmetric around the mean. 
- The left half is a mirror image of the right half

2. **Bell-shaped Curve**:

- The shape of the distribution is bell-shaped, with the highest point at the mean. 

3. **Mean, Median, and Mode**:

- In a normal distribution, the mean, median and mode are all equal and located at the center of the distribution. 

4. **Empirical Rule**:

- Approximately 68% of the data falls within one standard deviation of the mean. 
- Approximately 95% falls within two standard deviations.
- Approximately 99.7% falls within three standard deviations. 

5. **Asymptotic**:

- The tails of the distribution approach the horizontal axis but never touch it. 

- The tails of a normal distribution represent the extreme values. They approach the horizontal axis asymptotically, meaning they never actually reach zero. The tails are important in assessing the likelihood of extreme events. 

6. **Kurtosis in the context of a normal distribution**:

- Kurtosis measures the "tailedness" of the distribution. In the context of a normal distribution:

    - A normal distribution has a kurtosis of 0 (mesokurtic)
    - Positive kurtosis (leptokurtic) indicates heavier tails and a sharper peak.
    - Negative kurtosis (platykurtic) indicates lighter tails and a flatter peak.


### Standard Normal Distribution and QQ Plots

The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1. 

Z-score

A z-score measures how many standard deviations an element is from the mean. It is calculated as:
z = (x‚àíŒº)/ œÉ
- z: z-score
- x: value
- Œº: mean of the distribution
- œÉ: standard deviation of the distribution
 
 - Z-scores are used to standardize values, allowing comparison across different normal distributions. 

**Warning**

Converting data to z-scores (standardizing or normalizing the data) does not make the data normally distributed. It just puts the data on the same scale as the standard normal distribution, often for comparison purposes. 

QQ-Plot 

- **QQ-Plot**: is use to visually determine how close a sample is to a specified distribution in this case, the normal distribution. The QQ-Plot orders the z-scores from low to high and plots each value‚Äôs zscore on the y-axis; the x-axis is the corresponding quantile of a normal distribution for that value‚Äôs rank.Since the data is normalized, the units correspond to the number of standard deviations away from the mean. If the points roughly fall on the diagonal line, then the sample distribution can be considered close to normal.


### Applications of Normal Distribution 

1. **Natural Phenomena**:  Many natural phenomena follow a normal distribution (e.g., heights, blood pressure, IQ scores).

2. **Measurement Errors**: Measurement errors are often normally distributed.

3. **Statistical Methods**: Many statistical methods assume normality due to its mathematical properties.

Examples of Normal Distribution 

- **Example 1: Heights of Adults**
The heights of adult men in a country are normally distributed with a mean of 70 inches and a standard deviation of 3 inches. We can use this information to calculate probabilities and make predictions about the population.

- **Example 2: IQ Scores**
IQ scores are designed to follow a normal distribution with a mean of 100 and a standard deviation of 15. This allows for comparison of individuals' scores relative to the general population.


### Limitations of Using the normal distribution

- Not all data follows a normal distribution. 
- Sensitive to outliers, which can distort the mean and standard deviation 
- Assumes symmetry, which may not be present in all datasets.

### How would you use a normal distribution to calculate probabilities?

- To calculate probabilities, you can use the cumulative distribution function (CDF) of the normal distribution. For example, to find the probability that a value ùë• is less than or equal to a certain value, you would calculate P(X‚â§x) using the CDF.

### How do you determine if a dataset is normally distributed?

- You can determine normality using visual methods (e.g., Q-Q plots, histograms) and statistical tests (e.g. Shapiro-Wilk test, Kolmogorov-Smirnov test). These methods help assess if the data follows a normal distribution.

### Process of transforming a non-normal distribution to a normal distribution

- Log transformation
- Square root transformation
- Box-Cox transformation
These transformations can stabilize variance and make the data more closely approximate a normal distribution.


### What are some common applications of the normal distribution in data science?
- Common applications include:
    - Hypothesis testing (e.g., t-tests)
    - Confidence intervals
    - Control charts in quality control
    - Predictive modeling (assuming normality of residuals)


### Confidence Intervals in the context of a normal distribution?

- A confidence interval provides a range of values within which the true population parameter is expected to lie with a certain level of confidence (e.g., 95%). For a normal distribution, a 95% confidence interval for the mean is given by: 

$$
\bar{x} \pm 1.96 \left( \frac{\sigma}{\sqrt{n}} \right)
$$

where is the sample mean, ùúé is the standard deviation, and ùëõ is the sample size.

### Central Limit Theorem 

The Central Limit Theorem states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original population's. provided the samples are independent and identically distributed with a finite mean and variance. 

Implications of CLT 

1. **Estimation of Population Parameters**: We can use the sample mean to estimate the population mean. 
2. **Hypothesis Testing**: Many hypothesis tests assume normality due to CLT. 
3. **Confidence Intervals**: Confidence intervals for population parameters can be constructed using the normal distribution. 

How does the Central Limit Theorem apply to normal distribution?
- The CLT implies that even if the population distribution is not normal, the distribution of the sample means will be approximately normal for sufficiently large sample sizes. This allows the use of normal distribution techniques for inference. 

Why is the Central Limit Theorem important in statistics?
- The CLT is important because it enables us to make inferences about population parameters using the normal distribution, even when population distribution is unknown or not normal. It underpins many statistical methods and hypothesis tests. 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Generate a normal distribution
np.random.seed(42)
mean = 50
std_dev = 10
data = np.random.normal(mean, std_dev, 10000)

# Plotting the distribution
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, bins=30, color='blue')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Calculate probabilities
z_score = (60 - mean) / std_dev
probability = norm.cdf(z_score)
print(f"Probability of value ‚â§ 60: {probability}")

# Demonstrate Central Limit Theorem
sample_means = []
sample_size = 30

for _ in range(1000):
    sample = np.random.choice(data, sample_size)
    sample_means.append(np.mean(sample))

# Plotting the distribution of sample means
plt.figure(figsize=(10, 6))
sns.histplot(sample_means, kde=True, bins=30, color='green')
plt.title('Distribution of Sample Means (Central Limit Theorem)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()

# Mean and standard deviation of sample means
mean_of_sample_means = np.mean(sample_means)
std_dev_of_sample_means = np.std(sample_means)

print(f"Mean of Sample Means: {mean_of_sample_means}")
print(f"Standard Deviation of Sample Means: {std_dev_of_sample_means}")
```

Problem 1: Percentage of Light Bulbs Lasting More than 1300 hours

Given: 
- Mean lifespan (ùúá) = 1200 hours 
- Standard deviation (œÉ) = 100 hours 
- Lifespans follow a normal distribution 

First questions says What percentage of light bulbs will last more than 1300 hours?

Step by Step Solution:
1. Calculate the z-score:
-    The Z-score is the number of standard deviations a data point is from the mean. 
- Formula for z-score:
$$
z = \frac{x - \mu}{\sigma}
$$

Here, 
ùë•=1300 hours, ùúá=1200 hours, and œÉ=100 hours.

$$
z = \frac{1300 - 1200}{100} = \frac{100}{100} = 1
$$

2. Find the probability:
- Use the Z-score to find the corresponding probability from the standard normal distribution table or a calculator.
- The Z-score of 1 corresponds to a cumulative probability of approximately 0.8413 (from the Z-table).

3. Calculate the Complement:
- The cumulative probability up to 1300 hours is 0.8413. 
- 
$$
P(X > 1300) = 1 - P(X \leq 1300) = 1 - 0.8413 = 0.1587
$$

Answer: Approximately 15.87% of the light bulbs will last more than 1300 hours.

Step by Step Solution

## Case Studies 

- Case Study 1: Retail Business Sales Forecasting
Scenario:
A retail business wants to forecast the sales for the upcoming holiday season. Historical sales data shows that the daily sales follow a normal distribution with a mean of $10,000 and a standard deviation of $2,500.

Questions:

- What is the probability that the sales on a particular day will exceed $12,500?
- The business wants to ensure that they have enough inventory to cover sales that exceed the 95th percentile of the daily sales distribution. What should their inventory level be?
- If the store wants to achieve a target sales of at least $15,000 on a given day, what is the probability that this target will be met?

```python
import scipy.stats as stats 

mean = 10000
std_dev = 2500 

# Problem 1: Probability that sales will excedd $12,500
x = 12500
z = (x - mean) / std_dev
probability_exceed = stats.norm.sf(z) # sf stands for the survival function, which is defined as 1‚àíCDF (Cumulative Distribution Function).

print(f"The probability that the sales will exceed $12,500 on a particular day is approximately {probability_exceed:.4f}")

#Problem 2: The business wants to ensure that they have enough inventory to cover sales that exceed the 95th percentile of the daily sales distribution. What should their inventory level be?
percentile = 0.95 
z_95th = stats.norm.ppf(percentile) # ppf stands for the percent point function, which is the inverse of the Cumulative Distribution Function (CDF). 
inventory_level = mean + z_95th * std_dev
print(f"The inventory level to cover sales exceeding the 95th percentile is approximately ${inventory_level:.2f}")


# Problem 3: If the store wants to achieve a target sales of at least $15,000 on a given day, what is the probability that this target will be met?
target_sales = 15000
z_target = (target_sales - mean) / std_dev
probability_target = stats.norm.sf(z_target)
print(f"The probability of achieving target sales of at least $15,000 on a given day is approximately {probability_target:.4f}")
```
