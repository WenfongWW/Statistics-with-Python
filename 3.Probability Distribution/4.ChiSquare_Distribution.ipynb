# Chi Square Distribution

The Chi-squared (œá¬≤) distribution is a a continuous probability distribution that is used many hypothesis tests. It is especially useful in hypothesis testing, particularly in tests of independence and goodness-of-fit tests, and in the estimation of population variances.

Definition 

The chi-square distribution with 
ùëò
k degrees of freedom is the distribution of a sum of the squares of 
ùëò
k independent standard normal random variables. If Z1, Z2, ... , Zk 
are independent standard normal variables (mean = 0 and variance = 1), then the random variable. 

$$
Q = Z_1^2 + Z_2^2 + \cdots + Z_k^2
$$

is distributed according to the chi-square distribution with ùëò degrees of freedom. This is denoted as
$$
Q \sim \chi^2(k)
$$

The chi-square statistics is calculated as: 

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

Where Oi are the observed frequencies, and Ei are the expected frequencies. 

## Properties 

1. **Mean and Variance**:
- **Mean**: The mean of a chi-square distribution with k degrees of freedom is k. 
- **Variance**: The variance is 2k.

2. **Skewness and kurtosis**: 
- **Skewness**: The skewness is  squareroot(8/k), which indicates that the distribution is skewed to the right for smaller k and becomes more symmetric as k increases. 
- **Kurtosis**: The excess kurtosis is 12/k. 

3. **Shape**: 
- The chi-square distribution is asymmetric and only defined for non-negative values (x‚â•0).
- For k = 2, the distribution is exponential. 
- As k increases, the distribution becomes more bell-shaped, approaching a normal distribution for large k. 

4. **Special Case**:
- When k = 1, the distribution is known as the chi distribution. 
- For k=2, it is identical to the exponential distribution with rate parameter 1/2. 





## The shape of chi-square distributions 

We can see how the shape of a chi-square distribution changes as the degrees of freedom (k) increase by looking at graphs of the chi-square probability density function. A probability density function is a function that describes a continuous probability distribution.

**When k is one or two**

When k is one or two, the chi-square distribution is a curve shaped like a backwards ‚ÄúJ.‚Äù The curve starts out high and then drops off, meaning that there is a high probability that Œß¬≤ is close to zero.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Degrees of freedom
dfs = [1]

# Create a range of x values
x = np.linspace(0, 20, 1000)

# Plot the PDF of the chi-square distribution for different degrees of freedom
plt.figure(figsize=(10, 6))
for df in dfs:
    plt.plot(x, chi2.pdf(x, df), label=f'df = {df}')

plt.title('Chi-Square Distribution - PDF')
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.grid(True)
plt.show()
```

When k is greater than two

When k is greater than two, the chi-square distribution is hump-shaped. The curve starts out low, increases, and then decreases again. There is low probability that Œß¬≤ is very close to or very far from zero. The most probable value of Œß¬≤ is Œß¬≤ ‚àí 2.

When k is only a bit greater than two, the distribution is much longer on the right side of its peak than its left (i.e., it is strongly right-skewed).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Degrees of freedom
dfs = [3]

# Create a range of x values
x = np.linspace(0, 20, 1000)

# Plot the PDF of the chi-square distribution for different degrees of freedom
plt.figure(figsize=(10, 6))
for df in dfs:
    plt.plot(x, chi2.pdf(x, df), label=f'df = {df}')

plt.title('Chi-Square Distribution - PDF')
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.grid(True)
plt.show()
```

As k increases, the distribution looks more and more similar to a normal distribution. In fact, when k is 30 or greater, a normal distribution is a good approximation of the chi-square distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2, norm

# Set degrees of freedom
dfs = [50]

# Generate x values
x = np.linspace(0, 80, 1000)

# Create plots for each df
plt.figure(figsize=(12, 8))
for df in dfs:
    # Calculate chi-square PDF
    plt.plot(x, chi2.pdf(x, df), label=f'Chi-square (df={df})')

# Plot a normal distribution for comparison
mean = dfs[-1]  # mean of the chi-square when df=50
variance = 2 * dfs[-1]
std_dev = np.sqrt(variance)
plt.plot(x, norm.pdf(x, mean, std_dev), label='Normal approximation', linestyle='dashed')

plt.title('Chi-Square Distribution and Normal Approximation')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

## Applications  

1. **Hypothesis Testing**
- The chi-square distribution is most commonly used in chi-square tests for goodness of fit of an observed distribution to a theoretical one, the independence of two criteria of classification of qualitative data, and in confidence interval estimation for a population standard deviation of a normal distribution from a sample standard deviation.
- Common tests that use the chi-square distribution include the Pearson's chi-square test and the likelihood-ratio chi-square test.

2. **Confidence Intervals**
- It is used to construct confidence intervals for variance in cases where the underlying distribution is normal.

3. **Test for Variance**:
- In tests involving the variance of a normal population, the ratio of the sample variance to the true variance follows a scaled chi-square distribution
