# Uniform Distribution: Overview 

Uniform distribution is one of the simplest probability distributions. There are two types of uniform distributions: discrete uniform distribution and continuous uniform distribution. 

Definition: A uniform distribution is a type of probability distribution in which all outcomes are equally likely. Each interval of values has the same probability.Also known as rectangular distribution because of its shape.

### **Discrete Uniform Distribution**
Definition: 
- A discrete uniform distribution is one where a finite number of values are equally likely to be observed. Every value in the set has the same probability. 

Properties:
- **Finite set of values**: The set of possible outcomes is finite. 
- **Equal Probabilities**: Each outcome has an equal probability of occurring. 

Probability Mass Function (PMF):
- For a discrete uniform distribution over n outcomes x1, x2,... xn, the PMF is given by:

$$
P(X = x_i) = \frac{1}{n} \quad \text{for } i = 1, 2, \ldots, n
$$

Example: 
- Rolling a fair six-sided die: Each face (1, 2, 3, 4, 5, 6) has an equal probability of 1/6. 

### **Continuous Uniform Distribution**

- Definition:
- A continuous uniform distribution is one where any value within a specified range is equally likely to occur. 


**Properties**:

- **Interval [ùëé,ùëè]**: The possible outcomes are within a continuous range from ùëé to b.
- **Equal Likelihood**: Any subinterval of equal length within [ùëé,ùëè] has the same probability.

**Probability Density Function (PDF)**:
- The PDF of a continuous random variable is a function that describes the likelihood of the random variable taking on a particular value. 
- The value of the PDF at any given point can be interpreted as providing a relative likelihood of the random variable being near that point. 

- For a continuous uniform distribution over the interval [a.b], the PDF is given by:
$$
f(x) = \begin{cases} 
\frac{1}{b - a} & \text{for } a \leq x \leq b \\
0 & \text{otherwise}
\end{cases}
$$


**Cumulative Distribution Function (CDF)**:
- The CDF of a random variable X is a function that gives the probability that X will take a value less than or equal to x. 
- CDF: Range 0 to 1.
- The CDF for the continuous uniform distribution is:
$$
F(x) = \begin{cases} 
0 & \text{for } x < a \\
\frac{x - a}{b - a} & \text{for } a \leq x \leq b \\
1 & \text{for } x > b
\end{cases}
$$

**Mean and Variance**:
- Mean :
$$
\mu = \frac{a + b}{2}
$$

- Variance:
$$
\sigma^2 = \frac{(b - a)^2}{12}
$$

Example:
- The time a bus arrives at a bus stop is uniformly distributed between 0 and 20 minutes. Any time between 0 and 20 minutes is equally likely.



## Example Using Python 

PDF and CDF of a Continuous Uniform Distribution
Let's visualize the PDF and CDF of a continuous uniform distribution with 
ùëé= 0 and b=10

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters
a = 0
b = 10

# Generate data
x = np.linspace(a - 2, b + 2, 1000)
pdf = stats.uniform.pdf(x, loc=a, scale=b-a)
cdf = stats.uniform.cdf(x, loc=a, scale=b-a)

# Plot PDF
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, pdf, 'b', label='PDF')
plt.fill_between(x, pdf, alpha=0.2, color='blue')
plt.title('Uniform Distribution PDF')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, 'r', label='CDF')
plt.title('Uniform Distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()

plt.tight_layout()
plt.show()
```

### Key concepts and Applications 

1. **Uniform Random Variables**:

- Uniform random variables are used in simulations and modeling when each outcome in a range is equally likely.

2. **Generating Random Numbers**:

- Uniform distribution is often used to generate random numbers within a specified range. In many programming languages, functions for random number generation use uniform distribution as a base.

3. **Applications in Real Life**:

- Random Sampling: Selecting a random sample from a population.
- Simulation: Monte Carlo simulations often start with uniform random variables.
- Random Time Intervals: Modeling scenarios where events occur at any time within a given interval with equal likelihood.


### Practical Example in Python
- Let's calculate and visualize some properties of a continuous uniform distribution.

Example: Bus Arrival Time
- Suppose a bus arrives at a bus stop at a random time uniformly distributed between 0 and 20 minutes.
1. Calculate the Probability:

- What is the probability that the bus arrives between 5 and 15 minutes?
2. Simulate and Plot:

- Simulate 1000 random bus arrival times and plot the histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


a = 0
b = 20

# Calculate the probability that the bus arrives between 5 and 15 minutes
prob_5_to_15 = stats.uniform.cdf(15, loc=a, scale=b-a) - stats.uniform.cdf(5, loc=a, scale=b-a)
print(f"Probability that the bus arrives between 5 and 15 minutes: {prob_5_to_15:.4f}")

# Simulate 1000 random bus arrival times
np.random.seed(42)
arrival_times = np.random.uniform(a, b, 1000)

plt.figure(figsize=(10, 5))
plt.hist(arrival_times, bins=20, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Histogram of Bus Arrival Times')
plt.xlabel('Minutes')
plt.ylabel('Density')

# Overlay PDF
x = np.linspace(a, b, 1000)
pdf = stats.uniform.pdf(x, loc=a, scale=b-a)
plt.plot(x, pdf, 'r-', lw=2, label='Uniform PDF')
plt.legend()

plt.show()
```
