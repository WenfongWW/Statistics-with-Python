# Student's T Distribution: Overview 

The Student's t-distribution, commonly known as the t-distribution, is a probability distribution that is used to estimate population parameters when the sample size is small and/or when the population variance is unknown. It is particularly useful in hypothesis testing and constructing confidence intervals for means.

## Key Characteristics
- **Bell-shaped and Symmetric**: Similar to the normal distribution, but with heavier tails.
- **Degrees of Freedom (df)**: Determines the shape of the distribution. As the degrees of freedom increase, the t-distribution approaches the normal distribution.
- **Mean**: The mean of the t-distribution is 0.
- **Variance**: Greater than 1 but approaches 1 as the degrees of freedom increase.

## Probability Density Function (PDF)
The PDF of the t-distribution with v degrees of freedom is given by:

$$
f(t|\nu) = \frac{\Gamma\left(\frac{\nu+1}{2}\right)}{\sqrt{\nu\pi}\,\Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu+1}{2}}
$$

where:
- v is the degrees of freedom. 

- Œì is the gamma function.

### Cumulative Distribution Function (CDF)

The CDF of the t-distribution cannot be expressed in a simple closed form, but it can be computed using numerical methods or looked up in t-tables.

### Moments

$$
Mean: 0 \quad \text{(for } \nu > 1\text{)}
$$


$$
Variance: \frac{\nu}{\nu - 2} \quad \text{(for } \nu > 2\text{)}. \quad \text{For } 1 < \nu \leq 2, \text{ the variance is infinite.}
$$

$$
Skewness: 0 \quad \text{(symmetrical distribution)}
$$


$$
Kurtosis: \frac{6}{\nu - 4} \quad \text{for } \nu > 4. \quad \text{Higher than normal distribution (leptokurtic).}
$$


### Relationship with Normal Distribution
- For large sample sizes, the t-distribution approximates the normal distribution. Specifically, as ŒΩ approaches infinity, the t-distribution converges to the standard normal distribution.
- For ùúà=1, the t-distribution is the same as the Cauchy distribution.

### Use Cases

1. **Hypothesis Testing**:

- Used in t-tests to determine if there is a significant difference between sample means or a sample mean and a known value.
- One-sample t-test, two-sample t-test (independent and paired), and others.

2. **Confidence Intervals**:
- Constructing confidence intervals for the population mean when the sample size is small or the population variance is unknown.


### Example Calculations 

One-Sample t-test 

To test if a sample mean is S
$$
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
$$

Where:

- xÀâ is the sample mean.
- ùúá0 is the known value to test against.
- ùë† is the sample standard deviation.
- ùëõ is the sample size.

Example: 

Suppose you have a sample of 10 observations with a mean of 5 and a standard deviation of 2. Test if the sample mean is significantly different from 0 at a 5% significance level.

Solution:

- Sample mean (-x) = 5
- Hypothesized mean (Œº0) = 0 
- Sample standard deviation (s) = 2 
- Sample size (n) = 10 

$$
t = \frac{5 - 0}{2 / \sqrt{10}} = \frac{5}{0.6325} \approx 7.9057
$$

**Confidence Intervals**
To construct a confidence interval for the population mean using the t-distribution:


$$
\bar{x} \pm t_{\alpha/2, \nu} \frac{s}{\sqrt{n}}
$$

Example:
Suppose you have a sample of 10 observations with a mean of 5 and a standard deviation of 2. Construct a 95% confidence interval for the population mean.

Solution:

- Sample mean (xÀâ) = 5
- Sample standard deviation (s) = 2
- Sample size (ùëõ) = 10
- Degrees of freedom (ùúà) = 9

Using a t-table, the critical t-value for 9 degrees of freedom at the 95% confidence level (2.5% in each tail) is approximately 2.262.

$$
\text{Margin of error} = t_{\alpha/2, \nu} \frac{s}{\sqrt{n}} = 2.262 \times \frac{2}{\sqrt{10}} = 2.262 \times 0.6325 \approx 1.43
$$

##  PDF of t-Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the range of t-values
x = np.linspace(-5, 5, 1000)

# Define degrees of freedom
df1 = 1
df5 = 5
df10 = 10
df30 = 30

# Calculate the t-distribution PDF for different degrees of freedom
y1 = stats.t.pdf(x, df1)
y5 = stats.t.pdf(x, df5)
y10 = stats.t.pdf(x, df10)
y30 = stats.t.pdf(x, df30)

# Plot the PDFs
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='df=1')
plt.plot(x, y5, label='df=5')
plt.plot(x, y10, label='df=10')
plt.plot(x, y30, label='df=30')
plt.xlabel('t')
plt.ylabel('Probability Density')
plt.title('t-Distribution PDFs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the range of t-values
x = np.linspace(-5, 5, 1000)

# Define degrees of freedom
df1 = 1
df5 = 5
df10 = 10
df30 = 30

# Calculate the t-distribution CDF for different degrees of freedom
y1 = stats.t.cdf(x, df1)
y5 = stats.t.cdf(x, df5)
y10 = stats.t.cdf(x, df10)
y30 = stats.t.cdf(x, df30)

# Plot the CDFs
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='df=1')
plt.plot(x, y5, label='df=5')
plt.plot(x, y10, label='df=10')
plt.plot(x, y30, label='df=30')
plt.xlabel('t')
plt.ylabel('Probability Density')
plt.title('t-Distribution CDFs')
plt.legend()
plt.grid(True)
plt.show()
```

## Practical Questions 

1. **How do you interpret the t-value in a t-test?**
- The t-value indicates how many standard deviations the sample mean is away from the hypothesized mean. A higher absolute t-value indicates a more significant difference.

2. **When would you use a t-test instead of a z-test?**
- Use a t-test when the sample size is small (typically n<30) and/or the population variance is unknown(œÉ^2) .
- The unknown population variance is more common scenario because we typically estimate the population variance using the sample variance (s^2)

3. **Explain the importance of degrees of freedom in the t-distribution.**
- Degress of freedom affect the shape of the t-distribution. Lower degrees of freedom result in heavier tails, indicating more variability. 

4. **How do you calculate the p-value in a t-test?**

- The p-value is calculated based on the t-statistic and the degrees of freedom. It represents the probability of observing the data, or something more extreme, under the null hypothesis.

5. **What assumptions are made when using the t-distribution?**

- The sample data are independent.
- The data are approximately normally distributed.
- The variance of the population is unknown and estimated from the sample.

## Pratical Steps to Determine the Use of t-Distribution

1. **Sample Data Collection**:

- Collect a sample and calculate the sample mean (ùë•Àâ) and sample standard deviation (ùë†).
2. **Assess the Availability of Population Variance**:

- Determine whether the population variance (œÉ^2) is known or unknown.
- If the population variance is known, use the z-test.
- If the population variance is unknown (which is often the case), use the t-test.

3. **Check Sample Size**:

- For small sample sizes (typically n<30), the t-distribution is used because the Central Limit Theorem does not sufficiently ensure normality.
- For larger sample sizes (n‚â•30), the t-distribution can still be used, but the z-test can also be applied due to the normal approximation.

## Case Study: Effectiveness of a New Drug 

Scenario:
A pharmaceutical company has developed a new drug intended to lower blood pressure. The company conducted a clinical trial with 15 patients to test the drug's effectiveness. The patients' blood pressure levels were measured before and after taking the drug. The mean reduction in blood pressure was found to be 8 mmHg with a standard deviation of 3 mmHg.

The company wants to know if the drug has a significant effect on reducing blood pressure. Assume the blood pressure reductions are normally distributed.

Questions
1. **Describe the problem and explain why the Student's t-distribution is appropriate in this scenario.**
- Answer: The problem involves testing whether a new drug has a significant effect on reducing blood pressure based on a small sample size (15 patients). The population variance is unknown. The Student's t-distribution is appropriate because it accounts for the additional uncertainty introduced by estimating the population variance from a small sample size.

2. **Formulate the null and alternative hypotheses to test if the drug has a significant effect on blood pressure.**
- Answer: Null hypothesis (H0): The drug has no effect on blood pressure (Œº=0).
- Alternative hypothesis (H1): The drug has a significant effect on blood pressure (ùúá‚â†0).

3. **Calculate the t-statistic for the sample data.**

Answer:
Given:

- Sample mean (Àâx) = 8 mmHg
- Hypothesized mean (ùúá0) = 0 mmHg
- Sample standard deviation (s) = 3 mmHg
- Sample size (n) = 15
$$
t = \frac{\bar{x} - \mu_0}{\frac{s}{\sqrt{n}}} = \frac{8 - 0}{\frac{3}{\sqrt{15}}} = \frac{8}{0.7746} \approx 10.33
$$


4. **Determine the critical t-value for a 5% significance level (two-tailed test) and compare it with the calculated t-statistic.**

Answer:
Degrees of freedom (v) = n‚àí1 = 15‚àí1=14.

Using a t-table, the critical t-value for ŒΩ=14 at a 5% significance level (two-tailed) is approximately ¬±2.145.

Since ‚à£10.33‚à£>2.145, we reject the null hypothesis.


5. **Explain how to interpret the p-value in the context of this hypothesis test.**

Answer:
The p-value represents the probability of observing the sample data, or something more extreme, assuming the null hypothesis is true. A small p-value (typically less than 0.05) indicates strong evidence against the null hypothesis. In this case, the p-value would be very small, providing strong evidence that the drug has a significant effect on blood pressure.

6. **Construct a 95% confidence interval for the mean reduction in blood pressure.**
Answer:
- The 95% confidence interval is calculated using the formula:

$$
\bar{x} \pm t_{\alpha/2, \nu} \frac{s}{\sqrt{n}}
$$

The margin of error is:
$$
2.145 \times \frac{3}{\sqrt{15}} = 2.145 \times 0.7746 \approx 1.66
$$

So, the confidence interval is:
$$
8 \pm \ 1.66 = (6.34, 9.66)
$$

7. **Explain the assumptions that need to be satisfied for the validity of the t-test in this scenario.**

Answer:
The assumptions for the t-test are:

- The sample data are independent.
- The data are approximately normally distributed.
- The population variance is unknown and is estimated by the sample variance.
