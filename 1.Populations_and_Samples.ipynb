# Population and Sample: Complete Overview 

### Definition of Population 

A population refers to the entire group of individuals or instances about whom we hope to learn. It includes all possible observations of a specific group. 

Examples:

- All citizens of a country.
- All students in a school.
- All the cars manufactured by a company in a year.

### Definition of Sample

A sample is a subset of the population, selected to represent the population in statistical analysis. The sample should be representative of the population to make valid inferences.

Examples:

- A group of 1000 citizens selected from a country.
- A group of 50 students selected from a school.
- A group of 200 cars selected from the total production of a company in a year.

### Importance of Sampling 

- **Cost and Time Efficiency**: Studying a sample is more feasible than studying the entire population.
- **Practicality**: It‚Äôs often impossible to collect data from an entire population.
- **Manageability**: Data analysis and interpretation are easier with a sample.

### Sampling Methods 

1. **Simple Random Sampling**:
Every member of the population has an equal chance of being selected.

- Example: Suppose you have a population of 1000 employees in a company. To select a simple random sample of 100 employees, you can write each employee‚Äôs name on a separate piece of paper, put all the pieces in a hat, and draw out 100 names.

2. **Stratified Sampling**:
The population is divided into subgroups (strata) based on shared characteristics, and samples are taken from each stratum.

- Example: Assume a high school has 1000 students divided into four grades: 250 freshmen, 250 sophomores, 250 juniors, and 250 seniors. If you want to sample 100 students, you would sample 25 students from each grade to ensure representation across all grades.

3. **Cluster Sampling**:
The population is divided into clusters, some of which are randomly selected, and all members of selected clusters are sampled.

- Example: Consider a city with 100 schools. You want to survey 500 students. Instead of sampling students from each school, you randomly select 10 schools (clusters) and survey all the students in these 10 schools.

4. **Systematic Sampling**:
Every nth member of the population is selected.

- Example: Imagine a library with 10,000 books. To sample 1000 books, you can randomly choose a starting point (say, the 5th book) and then select every 10th book (5th, 15th, 25th, ...).

### Parameters and Statistics 

- **Parameter**: A characteristic or measure obtained by using all the data points in a population.

- **Statistic**: A characteristic or measure obtained by using the data points in a sample.


Examples of Parameters and Statistics

- **Population Mean (ùúá)**: The average of all data points in the population.
- **Sample Mean (ùë•-)**: The average of data points in a sample.
- **Population Proportion (P)**: The proportion of a characteristic in the entire population.
- **Sample Proportion (p)**: The proportion of a characteristic in the sample.

## Example in Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Simulate a population Dataset

```python
np.random.seed(42)
population_size = 10000
population_data = np.random.normal(loc=50, scale=15, size=population_size)
population_df = pd.DataFrame(population_data, columns=['value'])
population_df
```

We simulate a population of 10,000 data points with a normal distribution (mean = 50, standard deviation = 15).

2. Draw a Sample from the Population:



```python
sample_size = 1000
sample_data = np.random.choice(population_data, size=sample_size, replace=False)
sample_df = pd.DataFrame(sample_data, columns=['value'])
sample_df
```

3. Calculate Statistics for Population and Sample:



```python
population_mean = population_df['value'].mean()
sample_mean = sample_df['value'].mean()
population_std = population_df['value'].std()
sample_std = sample_df['value'].std()

print(f"Population Mean: {population_mean}")
print(f"Sample Mean: {sample_mean}")
print(f"Population Standard Deviation: {population_std}")
print(f"Sample Standard Deviation: {sample_std}")
```

4. Visualization

```python
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
sns.histplot(population_df['value'], kde=True, color='blue', bins=30)
plt.title('Population Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(sample_df['value'], kde=True, color='orange', bins=30)
plt.title('Sample Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
